## 🔧 2. 加载配置文件

从项目根目录的 `config.json` 读取模型配置。

```python
import json
from pathlib import Path

# 加载配置文件（使用当前工作目录）
config_path = Path.cwd() / ".." / "config.json"

with open(config_path, "r", encoding="utf-8") as f:
    config = json.load(f)

# 提取模型配置
MODEL_CONFIG = config["model"]

print("✅ 配置加载成功")
print(f"   提供商：{MODEL_CONFIG['provider']}")
print(f"   模型：{MODEL_CONFIG['name']}")
print(f"   地址：{MODEL_CONFIG['base_url']}")
print(f"   最大 token: {MODEL_CONFIG['max_tokens']}")
```

---

## 🚀 3. 定义调用函数

```python
import requests

def call_llm(prompt: str) -> str:
    """
    调用 LLM 模型
    
    Args:
        prompt: 用户输入
    
    Returns:
        模型回复
    """
    url = f"{MODEL_CONFIG['base_url']}/v1/chat/completions"
    
    payload = {
        "model": MODEL_CONFIG['name'],
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "temperature": MODEL_CONFIG.get('temperature', 0.7),
        "max_tokens": MODEL_CONFIG.get('max_tokens', 120000)
    }
    
    try:
        response = requests.post(url, json=payload, timeout=120)
        response.raise_for_status()
        result = response.json()
        return result["choices"][0]["message"]["content"]
    except requests.exceptions.ConnectionError:
        return f"❌ 错误：无法连接到模型服务 ({MODEL_CONFIG['base_url']})"
    except Exception as e:
        return f"❌ 错误：{str(e)}"

print("✅ 调用函数定义完成")
```

---

## 🎮 4. 测试调用

```python
# 测试调用
prompt = "你好，请介绍一下你自己"
print(f"Q: {prompt}")
answer = call_llm(prompt)
print(f"A: {answer}")
```

```python
# 更多测试
test_prompts = [
    "1 + 1 等于多少？",
    "用一句话描述春天",
    "什么是量子力学？",
]

for prompt in test_prompts:
    print(f"\n{'='*50}")
    print(f"Q: {prompt}")
    answer = call_llm(prompt)
    print(f"A: {answer}")
```

## 🎯 5. 练习：尝试不同的 Prompt

```python
# 练习 1：让模型扮演诗人角色
prompt = "请以诗人的身份，写一句关于月亮的话"
print(f"Q: {prompt}")
answer = call_llm(prompt)
print(f"A: {answer}")
```

```python
# 练习 2：让模型扮演程序员角色
prompt = "请以程序员的身份，用递归的例子解释递归"
print(f"Q: {prompt}")
answer = call_llm(prompt)
print(f"A: {answer}")
```

```python
# 练习 3：添加系统提示

def call_llm_with_system(prompt: str, system_prompt: str = "你是一个有帮助的助手") -> str:
    """
    使用系统提示调用模型
    """
    url = f"{MODEL_CONFIG['base_url']}/v1/chat/completions"
    
    payload = {
        "model": MODEL_CONFIG['name'],
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ],
        "temperature": MODEL_CONFIG.get('temperature', 0.7),
        "max_tokens": MODEL_CONFIG.get('max_tokens', 120000)
    }
    
    response = requests.post(url, json=payload, timeout=120)
    result = response.json()
    return result["choices"][0]["message"]["content"]

# 测试
prompt = "什么是量子力学？"
system = "你是一位物理学教授，用通俗易懂的方式解释"
print(f"系统提示：{system}")
print(f"Q: {prompt}")
answer = call_llm_with_system(prompt, system)
print(f"A: {answer}")
```
